```python
from gensim.test.utils import datapath
from gensim import utils
import gensim

class MyCorpus(object):
    """An interator that yields sentences (lists of str)."""

    def __iter__(self):
        corpus_path = datapath('/Users/Palma/Documents/Githubs/v4w_website/app/static/files/lista_key.txt')
        for line in open(corpus_path):
            # assume there's one document per line, tokens separated by whitespace
            yield utils.simple_preprocess(line)
            
            
```

```python
lista_completa = MyCorpus()
```

```python
for posto in lista_completa:
    print(posto)
```

```python
model = gensim.models.Word2Vec(sentences=lista_completa)
```

```python
model.wv['ponte de ca rizzi']
```

```python
import numpy as np
from scipy import spatial

index2word_set = set(model.wv.index2word)

def avg_feature_vector(sentence, model, num_features, index2word_set):
    words = sentence.split()
    feature_vec = np.zeros((num_features, ), dtype='float32')
    n_words = 0
    for word in words:
        if word in index2word_set:
            n_words += 1
            feature_vec = np.add(feature_vec, model[word])
    if (n_words > 0):
        feature_vec = np.divide(feature_vec, n_words)
    return feature_vec
```

```python
s1_afv = avg_feature_vector('rialto', model=model, num_features=100, index2word_set=index2word_set)
s2_afv = avg_feature_vector('castello', model=model, num_features=100, index2word_set=index2word_set)
sim = 1 - spatial.distance.cosine(s1_afv, s2_afv)
print(sim)
```
